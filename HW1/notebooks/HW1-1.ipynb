## 1.1 Definition of rank of a homography

### Q1.1

Size of $S$ will depend on the dimensions of $p$ and $q$.

- $p$ has $N$ dimension + 1 for bias term for projective coordinates.
- $q$ has $M$ dimension + 1 for bias term for projective coordinates.

S will have size ($N+1$, $M+1$) (row, column ordering)

---

Say that points $p$ and $q$ are both 2D points in the form:

$$p = \begin{bmatrix}x \\ y \\ 1 \end{bmatrix}$$
$$q = \begin{bmatrix}x^{'} \\ y^{'} \\ 1 \end{bmatrix}$$

S would take be a $3x3$ matrix.

$$
S = \begin{bmatrix}
s_1 & s_2 & s_3 \\
s_4 & s_5 & s_6 \\
s_7 & s_8 & s_9
\end{bmatrix}
$$

For brevity sake, we can substitude in our values into the provided equation (1) to form:

$$
p \equiv S q
$$

$$
\begin{bmatrix}x \\ y \\ 1 \end{bmatrix}
\equiv
\begin{bmatrix}
s_1 & s_2 & s_3 \\
s_4 & s_5 & s_6 \\
s_7 & s_8 & s_9
\end{bmatrix}
\begin{bmatrix}x^{'} \\ y^{'} \\ 1 \end{bmatrix}
$$

---

### Q1.2

The rank of $S$ depends on the dimensions of $p$ and $q$ as well as the degrees of freedom between each view. The rank of the matrix described above is 3 (full rank matrix) since there are no constraints in either view.

---

### Q1.3

4 points of correspondence are required to estimate S with the requirement that no point is co-linear to any other three points.

When solving for homography matrix $S$, 8 values of $h$ to solve for (1 element is set to 1),
2 knowns per point,
thus 4 points.

Any fewer than 4 and the system would be underdetermined.

## 1.2 Homography between cameras

### Q1.4

Put $u$ and $v$ in terms of $M_1$ and $M_2$ respectively.

$$
u = M_1 P
$$
$$
v= M_2 P
$$

Solving for v leads to the following equation:

$$
v = M_2 M_1^{-1} u
$$

We can refer our previously defined homography equation:

$$
p = Sq
$$

...which looks rather familar to our equation for $v$ above, thus there exists a homography between points u and v and the top left 3 x 3 matrix of $M_2 M_1^{-1}$, S (the 4th column is the translation between views and the 4th row is projection related variables).
