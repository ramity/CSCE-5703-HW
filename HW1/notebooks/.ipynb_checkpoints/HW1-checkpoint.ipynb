### Q2.3

```python
def normalize(x: np.array, nd: int = 2):
    
    # Perform some analysis on x
    
    mean = np.mean(x)
    std = np.std(x)

    
    # Populate W with obtained stats on x
    
    W = np.zeros(3,3)
    W[0][0] = std
    W[0][1] = 0
    W[0][2] = mean[0]
    W[1][0] = 0
    W[1][1] = std
    W[1][2] = mean[1]
    W[2][0] = 0
    W[2][1] = 0
    W[2][2] = 1

    # Invert W, perform calculation and return W + normalized x 
    
    W_inv = np.linalg.inv(W)
    x_ext = np.concatenate([x.T, np.ones((1, x.shape[0]))])
    normalized_x = np.dot(W_inv, x_ext)
    normalized_x = normalized_x[0:nd, :].T
    
    return W_inv, normalized_x
    

def computeH_norm(p1: np.array, p2: np.array):
    
    # Normalize points
    
    W1, normalized_p1 = normalize(p1, 2)
    W2, normalized_p2 = normalize(p2, 2)
    
    
    # Calculate W2^+
    
    W2_pinv = np.linalg.pinv(W2)
    
    
    # Pass normalized points to computH and return response
    
    H = computeH(normalized_p1, normalized_p2)
    
    
    # Perform some calculations to complete and return the calculation of h
    
    W2_pinv_dot_H = np.dot(W2_pinv, H)
    H = np.dot(W2_pinv_dot_H, W1)
    H = H / H[-1, -1]
    return H
```

### Q2.5

```python
def calcM(image_1, image_2, H):
    
    # Initialize needed vars
    
    h1, w1 = image_1.shape[:2]
    h2, w2 = image_2.shape[:2]
    pts1 = np.zeros(4, 2)
    pts2 = np.zeros(4, 2)
    
    
    # Define pts1 and pts2
    
    pts1[0][0] = 0
    pts1[0][1] = 0
    pts1[1][0] = 0
    pts1[1][1] = h1
    pts1[2][0] = w1
    pts1[2][1] = h1
    pts1[3][0] = w1
    pts1[3][1] = 0

    pts2[0][0] = 0
    pts2[0][1] = 0
    pts2[1][0] = 0
    pts2[1][1] = h2
    pts2[2][0] = w2
    pts2[2][1] = h2
    pts2[3][0] = w2
    pts2[3][1] = 0
    
    
    # Reshape and convert to floats for better accuracy
    
    pts1 = pts1.reshape(-1, 1, 2).astype(np.float32)
    pts2 = pts2.reshape(-1, 1, 2).astype(np.float32)
    
    
    # Define pts + steps to populate needed vars to populate M 
    
    pts2_ = cv2.perspectiveTransformation(pts2, H)
    pts = np.concatenate((pts1, pts2_), axis=0)
    [xmin, ymin] = np.array(pts.min(axis=0).ravel() - 0.5).astype(np.int32)
    [xmax, ymax] = np.array(pts.max(axis=0).ravel() + 0.5).astype(np.int32)
    scale = 1280 / (ymax - ymin)
    out_x = int(scale * (xmax - xmin))
    out_y = int(scale * (ymax - ymin))
    out_shape = (out_x, out_y)

    
    # Initialize, populate, and return M
    
    M = np.zeros(3,3)
    
    M[0][0] = scale
    M[0][1] = 0
    M[0][2] = -scale * xmin
    M[1][0] = 0
    M[1][1] = scale
    M[1][2] = -scale * ymin
    M[2][0] = 0
    M[2][1] = 0
    M[2][2] = 1

    return M, out_shape
```

### Q2.6

```python
def warpTwoImages(image_1, image_2, H):
    
    # Initialize needed vars
    
    h1, w1 = image_1.shape[:2]
    h2, w2 = image_2.shape[:2]
    pts1 = np.zeros(4, 2)
    pts2 = np.zeros(4, 2)
    
    
    # Define pts1 and pts2
    
    pts1[0][0] = 0
    pts1[0][1] = 0
    pts1[1][0] = 0
    pts1[1][1] = h1
    pts1[2][0] = w1
    pts1[2][1] = h1
    pts1[3][0] = w1
    pts1[3][1] = 0

    pts2[0][0] = 0
    pts2[0][1] = 0
    pts2[1][0] = 0
    pts2[1][1] = h2
    pts2[2][0] = w2
    pts2[2][1] = h2
    pts2[3][0] = w2
    pts2[3][1] = 0
    
    
    # Reshape and convert to floats for better accuracy
    
    pts1 = pts1.reshape(-1, 1, 2).astype(np.float32)
    pts2 = pts2.reshape(-1, 1, 2).astype(np.float32)
    
    
    # Define pts + steps to populate needed vars to populate M 
    
    pts2_ = cv2.perspectiveTransformation(pts2, H)
    pts = np.concatenate((pts1, pts2_), axis=0)
    [xmin, ymin] = np.array(pts.min(axis=0).ravel() - 0.5).astype(np.int32)
    [xmax, ymax] = np.array(pts.max(axis=0).ravel() + 0.5).astype(np.int32)
    scale = 1280 / (ymax - ymin)
    out_x = int(scale * (xmax - xmin))
    out_y = int(scale * (ymax - ymin))
    out_shape = (out_x, out_y)

    
    # Initialize, populate, and return M
    
    M = np.zeros(3,3)
    
    M[0][0] = scale
    M[0][1] = 0
    M[0][2] = -scale * xmin
    M[1][0] = 0
    M[1][1] = scale
    M[1][2] = -scale * ymin
    M[2][0] = 0
    M[2][1] = 0
    M[2][2] = 1
    
    result = cv2.warpPerspective(image_2, M.dot(H), out_shape)
    
    intermediate = np.zeros_like(result)
    intermediate[int(-scale * ymin) : int(-scale * ymin) + int(scale * h1),
                 int(-scale * xmin) : int(-scale * xmin) + int(scale * w1)] =\
                 cv2.resize(image_1, (int(scale * w1), int(scale * h1)))
    
    mask = np.logical_and(result > 0, intermediate > 0)
    blend = 0.5 * result[mask] + 0.5 * intermediate[mask]
    result[int(-scale * ymin) : int(-scale * ymin) + int(scale * h1),
           int(-scale * xmin) : int(-scale * xmin) + int(scale * w1)] =\
           cv2.resize(image_1, (int(scale * w1), int(scale * h1)))
    result[mask] = blend
    return result
```

### Q3.1

```python
def DLT(object_3d_coords, image_2d_coords):
    
    N = object_3d_coords.shape[0]
    
    
    # Make sure vars meet the following conditions
    
    assert N >= 6, "DLT requires @least 6 points"
    assert image_2d_coords.shape[0] == N
    assert object_3d_coords.shape[1] == 3
    
    
    T_object_3d_coords, object_3d_coords_n = normalize(object_3d_coords, 3)
    T_image_2d_coords, image_2d_coords_n = normalize(image_2d_coords, 2)
    
    
    A = []
    for i in range(N):
        x, y, z = object_3d_coords_n[i, 0], object_3d_coords_n[i, 1], object_3d_coords_n[i, 2]
        u, v = image_2d_coords_n[i, 0], image_2d_coords_n[i, 1]
        A.append([x, y, z, 1, 0, 0, 0, 0, -u * x, -u * y, -u * z, -u])
        A.append([0, 0, 0, 0, x, y, z, 1, -v * x, -v * y, -v * z, -v])
    
    
    A = np.asarray(A)
    U, S, V = np.linalg.svd(A)
    L = V[-1, :] / V[-1, -1]
    H = K.reshape(3, 4)
    
    
    # Unnormalize
    
    H = np.dot(np.dot(np.linalg.pinv(T_image_2d_coords), H), T_object_3d_coords)
    H = H / H[-1, -1]
    return H


def estimateCamera(image_2d_coords, object_3d_coords):
    
    P = DLT(object_3d_coords, image_2d_coords)
    
    return P[:, 0 : 3], P[:, 3]
```

### Q3.2

Rotation cannot be the first 3x3 part of M since M is the product of intrinsic and extrinsic parameters.

### Q3.3

If the intinsic parameters is the identity matrix, the first 3x3 part of M will be the rotation matrix because the following equation:

$$
M = K[R | t]
$$

Would decompose into:

$$
M = [R | t]
$$

### Q3.4

Use RQ decomposition to convert first 3x3 part of M into the rotation matrix.

### Q3.5

```python
from scipy.linalg import rq


def estimateCamera2(image_2d_coords, object_3d_coords):
    
    # Estimate camera matrix w/o projective row and estimate C
    
    P = DLT(object_3d_coords, image_2d_coords)
    U, S, V = np.linalg.svd(P)
    C = V[-1 , 0 : 3] / V[-1, -1]
    
    
    # Estimate K and R via RQ
    
    K, R = rq(P[0 : 3, 0 : 3])
    D = np.diag(np.sign(np.diag(K)))
    K = K @ D
    R = D @ R
    t = -R @ C
    return R, t
```
